```python
df = pd.read_excel('Main Dataset/Skewed dataset/Customer_with_features.xlsx')
df.replace(to_replace = np.nan, value =0, inplace=True)
df.head()
```

```python
#Assigning 2 new columns for Retained_Amount and Retained_Orders
a = df['Total_Amount']
b = df['Refund_Amount']
c = df['Total_Orders']
d = df['Total_Returns']

Actual_amount = a - b
Actual_orders = c - d
df = df.assign(Retained_Amount = Actual_amount)
df = df.assign(Retained_Orders = Actual_orders)
df3 = df
df3.head()
```

```python
df3.to_excel('Main Dataset/Skewed dataset/Customer_Shopping_Details.xlsx')
```

```python
#Normalization
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df2 = pd.DataFrame(x_scaled)
df2.columns = ['Customer ID','Total_Returns','Total_Orders','Total_amount','Refund_amount','Difference_in_days','Rating','Retained_Amount','Retained_Orders']
Customer_ID = df['Customer_ID']
df2['Customer_ID'] = Customer_ID
df2 = df2.drop(columns=['Customer ID'])
df2= df2[['Customer_ID','Retained_Orders','Total_Returns','Total_Orders','Retained_Amount','Total_amount','Refund_amount','Difference_in_days','Rating']]
df = df2

df.head()
```

```python
df.to_excel('Main Dataset/Skewed dataset/Normalised_dataset.xlsx')
```

```python
#Parameter Declaration
Retained_Amount_Parameter = 0.8
Retained_Orders_Parameter = 0.2
Refund_amount_Parameter = 0.3
Total_returns_Parameter = 0.65
Difference_in_days_Parameter = 0.03
Rating_Parameter = 0.02

#credit_score
credit_list = list()

# iterate over rows with iterrows()
for index, row in df.iterrows():
   
    #Column Names
    a = row['Retained_Amount']
    b = row['Retained_Orders']
    c = row['Refund_amount']
    d = row['Total_Returns']
    e = row['Difference_in_days']
    f = row['Rating']
    
#     print(index, row['Retained_Orders'], row['Retained_Amount'], row['Refund_amount'])
    
    credit_score= 500 + ((Retained_Amount_Parameter*a + Retained_Orders_Parameter* b) - (Refund_amount_Parameter*c
                + Total_returns_Parameter*d + Difference_in_days_Parameter*e + Rating_Parameter*f ))*500
#     print(credit_score)
    credit_list.append(credit_score)
# print(credit_list)
df = df.assign(Credit_Score = credit_list)
df.head()
```

```python
df.to_excel('Main Dataset/Skewed dataset/Credit_Dataset.xlsx')
```
